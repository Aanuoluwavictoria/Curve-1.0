# Implementation of Logistic Regression 

```python
# importing the librarice
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

# Improt the dataset and extract the independent and dpendent variables:

```python
#importing the dataset
social_network = pd.read_csv(r'C:\Users\Adeshola\Documents\Aanuoluwa data File\Soical Networking\Social_Network_Ads.csv')
social_network
```

```python
# Etrating the independent variables
x = social_network.iloc[:, [2,3]].values

#Extracting the dependent variables
y = social_network.iloc[:,4].values
```

```python
x
```

# Visualize the dataset

```python
sns.heatmap(social_network.corr())
```

# Split the data into traning and testing set

```python
#spliting the dataset into the traning set and testset
from sklearn.model_selection import train_test_split
x_test,x_train, y_test, y_train=train_test_split(x, y, test_size = 0.25,random_state = 0)
```

# Fit simple linear regression to the training dataset

```python
from sklearn.preprocessing import StandardScaler

sc_x = StandardScaler()  # Create an instance of StandardScaler
x_train = sc_x.fit_transform(x_train)  # Fit to data, then transform it
x_test = sc_x.transform(x_test)  # Transform the test data
```

# Fit Logistics Regression to Train the data set

```python
from sklearn.linear_model import LogisticRegression
logR=LogisticRegression(random_state=0)
logR.fit(x_train,y_train)
```

# Predict the test set result

```python
y_pred = logR.predict(x_test)
y_pred
```

# Visalize the Train set resultts:

```python
from matplotlib.colors import BoundaryNorm

# Assuming x_train and y_train are defined
x_set, y_set = x_train, y_train
x1, x2 = np.meshgrid(np.arange(start=x_set[:, 0].min() - 1, stop=x_set[:, 0].max() + 1, step=0.01),
                     np.arange(start=x_set[:, 1].min() - 1, stop=x_set[:, 1].max() + 1, step=0.01))

# Plot decision boundary
plt.contourf(x1, x2, logR.predict(np.array([x1.ravel(), x2.ravel()]).T).reshape(x1.shape),
             alpha=0.75, cmap=plt.cm.Paired)

plt.xlim(x1.min(), x1.max())
plt.ylim(x2.min(), x2.max())

for i, j in enumerate(np.unique(y_set)):
    plt.scatter(x_set[y_set == j, 0], x_set[y_set == j, 1],
                c=plt.cm.Paired(i / len(np.unique(y_set))), label=j)

plt.title('Logistic Regression (Training set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```

# Evaluating the model

```python
#Confussion  matrix to evaluate the model
#Evaluate the model using confusion matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)
```
